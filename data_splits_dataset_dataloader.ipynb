```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
import json
f = open('data/LDCTIQAG2023_train/train.json')
data = json.load(f)
f.close()
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
```

```python
df.columns = ['sc']
df.head(1)
```

```python
df['image_id'] = df.index
df['score'] = df['sc']
```

```python
df = df.drop(['sc'], axis=1)
df.head()
```

```python
df = df.reset_index().drop(['index'], axis=1)
df.head()
```

```python
df.shape
```

```python
n_samples = len(df)
```

```python
df_other, df_val1 = train_test_split(df, test_size=n_samples//5, random_state=0, stratify=df.score)
df_other, df_val2 = train_test_split(df_other, test_size=n_samples//5, random_state=0, stratify=df_other.score)
df_other, df_val3 = train_test_split(df_other, test_size=n_samples//5, random_state=0, stratify=df_other.score)
df_val4, df_val5 = train_test_split(df_other, test_size=n_samples//5, random_state=0, stratify=df_other.score)

df_train1 = pd.concat([df_val2, df_val3, df_val4, df_val5])
df_train2 = pd.concat([df_val1, df_val3, df_val4, df_val5])
df_train3 = pd.concat([df_val1, df_val2, df_val4, df_val5])
df_train4 = pd.concat([df_val1, df_val2, df_val3, df_val5])
df_train5 = pd.concat([df_val1, df_val2, df_val3, df_val4])

df_train1.to_csv('data/train_f1.csv', index=None)
df_val1.to_csv('data/val_f1.csv', index=None)

df_train2.to_csv('data/train_f2.csv', index=None)
df_val2.to_csv('data/val_f2.csv', index=None)

df_train3.to_csv('data/train_f3.csv', index=None)
df_val3.to_csv('data/val_f3.csv', index=None)

df_train4.to_csv('data/train_f4.csv', index=None)
df_val4.to_csv('data/val_f4.csv', index=None)

df_train5.to_csv('data/train_f5.csv', index=None)
df_val5.to_csv('data/val_f5.csv', index=None)
```
